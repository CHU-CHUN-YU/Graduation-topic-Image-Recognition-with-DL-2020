```python
#picture 1 vgg train vs test vs validation

plt.figure(figsize=(10,7))
test_vgg1 = list([0.898,0.890,0.918,0.939,0.948,0.929])
train_vgg1 = list([0.978,0.991,0.995,0.998,0.999,0.999])
val_vgg1 = list([0.961,0.977,0.981,0.983,0.987,0.982])

y = count
plt.plot(y, test_vgg1,'s-', color ='r',label='Test')
plt.plot(y, train_vgg1,'o-', color ='y',label='Train')
plt.plot(y, val_vgg1,'s-', color ='b',label='Validation')
plt.title('InceptionV3  Picture 04',fontsize=20)    # 當有使用 plt.subplot， 總 title
#plt.suptitle('Accuracy')              # 每一個plot()的title
plt.xlabel('epoch',fontsize=15)
plt.ylabel('accuracy',fontsize=15)
#plt.vlines(x=60,ymin=0.55,ymax=1)
plt.legend(loc = "lower right", fontsize=15)
for i in range(6):
    plt.text(y[i]-3, test_vgg1[i], test_vgg1[i], fontsize=12)

for i in range(6):
    plt.text(y[i] , train_vgg1[i], train_vgg1[i], fontsize=12)

for i in range(6):
    plt.text(y[i]-2, val_vgg1[i]+0.005, val_vgg1[i], fontsize=12)

plt.savefig(r'C:\Users\user\Desktop\project\image04\INCV3_train_test_p4 accuracy.jpg')
plt.show()
```
